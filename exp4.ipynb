```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
Gray_image = cv2.imread('pic3.jpg')
```

```python
Color_image = cv2.imread('pic1.jpg')
```

```python
plt.imshow(Gray_image)
plt.show()
```

```python
plt.imshow(Color_image)
plt.show()
```

```python
hist = cv2.calcHist([Gray_image],[0],None ,[256],[0, 256])
hist2 = cv2.calcHist([Color_image],[1], None, [256], [0, 2560])
```

```python
plt.figure()
plt.title("Histogram")
plt.xlabel('grayscale value')
plt.ylabel('pixel count')
plt.stem(hist)
plt.show()
```

```python
plt.figure()
plt.title("Histogram")
plt.xlabel('grayscale value')
plt.ylabel('pixel count')
plt.stem(hist2)
plt.show()
```
